```python
def plot_regression_analysis(hrs_estudio, cali, grupo):
    X = np.array(hrs_estudio).reshape(-1, 1)
    y = np.array(cali)

    modelo = LinearRegression()

    modelo.fit(X, y)

    predicciones = modelo.predict(X)

    mse = mean_squared_error(y, predicciones)
    r2 = r2_score(y, predicciones)

    print(f"Grupo {grupo}:")
    print(f"Coeficiente: {modelo.coef_[0]}")
    print(f"Intercepto: {modelo.intercept_}")
    print(f"MSE: {mse}")
    print(f"R^2: {r2}\n")

    plt.scatter(hrs_estudio, cali, color='blue', label='Datos reales')
    plt.plot(hrs_estudio, predicciones, color='red', label='Regresión lineal')
    plt.xlabel('Horas de estudio')
    plt.ylabel('Calificaciones')
    plt.title(f'Grupo {grupo} - Regresión Lineal')
    plt.legend()
    plt.show()

plot_regression_analysis(hrs_estudio_grupo1, cali_grupo1, 1)
plot_regression_analysis(hrs_estudio_grupo2, cali_grupo2, 2)
plot_regression_analysis(hrs_estudio_grupo3, cali_grupo3, 3)
```
